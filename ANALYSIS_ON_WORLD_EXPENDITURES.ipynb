```python
custom_palette = ["deeppink", "skyblue", "yellow", "green"]
sns.barplot(x="Country",y="GDP(%)",data=df.head(700),palette=custom_palette)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
sns.barplot(x="Country",y="Expenditure(million USD)",data=df.tail(640),palette="Blues")
plt.tight_layout()
plt.xticks(rotation=90)
```

```python
sns.scatterplot(x="Sector",y="GDP(%)",data=df.sample(320))
plt.xticks(rotation=90)
plt.title("sector and GDP(%)")
```

```python
sns.boxplot(x="Sector",y="Country",data=df.head(400))
plt.xticks(rotation=90)
plt.yticks(rotation=360)
```

```python
plt.figure(figsize=(12,6))
sns.pointplot(x="Country",y="GDP(%)",data=df.sample(500))
plt.xticks(rotation=90)
```

```python
sns.jointplot(x="Expenditure(million USD)",data=df.head(420),kind="kde")
```

```python

plt.fill_between(df["Year"], df["GDP(%)"], color="skyblue", alpha=0.4)
plt.plot(df["Year"], df["GDP(%)"], color="Slateblue", alpha=0.6, linewidth=2)

plt.xlabel("Year")
plt.ylabel("GDP(%)")
plt.title("GDP Percentage Over Years")
plt.grid(True)
plt.show()
```
