```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv("/content/engine size and its price2.csv")
df
```

```python
df.drop(["vehicle"],axis=1)
```

```python
df.drop(["top"],axis=1)
```

```python
#bar plot
x=df["engine size"].head(3)
y=df["price"].head(3)
plt.xlabel("Engine size",fontsize=13,color='blue')
plt.ylabel("Price",fontsize=13,color='red')
plt.xticks(color='red',fontsize=16)
plt.yticks(color="blue",fontsize=16)
plt.title("Car price",fontsize=29,color="red",alpha=1,)
plt.tight_layout()
plt.legend(['price'])
plt.bar(x,y,label="Price",color='blue')
```

```python
#line plot
x=df["engine size"]
y=df["price"]
plt.xlabel("Engine size",fontsize=16,color='deeppink')
plt.ylabel("Price",fontsize=16,color='deeppink')
plt.xticks(color='yellowgreen',fontsize=16)
plt.yticks(color="yellowgreen",fontsize=16)
plt.title("Car price",fontsize=29,color="yellowgreen",alpha=1,)
plt.tight_layout()
plt.plot(x,y,label="Price",color='deeppink',alpha=1)
```

```python
#scatter plot
x=df["horsepower"]
y=df["price"]
plt.xlabel("Horse Power",fontsize=16,color='grey')
plt.ylabel("Price",fontsize=16,color='grey')
plt.xticks(color='grey',fontsize=16)
plt.yticks(color="grey",fontsize=16)
plt.title("Car price",fontsize=29,color="grey",alpha=1,)
plt.tight_layout()
plt.scatter(x,y,label="Price",color='yellow',alpha=1)
```

```python
#histogram
y=df["price"]
plt.hist(y,color='green',alpha=1)
```

```python
#Area plot
x = df["engine size"]
y = df["price"]
plt.xlabel("Engine size", fontsize=16, color='purple')
plt.ylabel("Price", fontsize=16, color='purple')
plt.xticks( fontsize=16)
plt.yticks( fontsize=16)
plt.title("Car Price vs. Engine Size", fontsize=20,color='purple' )
plt.tight_layout()
plt.fill_between(x, y, color='violet', alpha=1)
```
